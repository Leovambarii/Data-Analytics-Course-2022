## Exercise 3
a) Create boxplot of apartment's grade versus it's price. Compare with plot 2b).  
b) Create mosaic plot of any two categorical variables.

```python
#a)
#Your code:
df_list = []    # List that will contain data of prices for each grade
labels_list = []        # List that will store the grades labels
grouped_df = df.groupby("grade")        # Grouping the data accorting to the grades
for grade in grouped_df.groups.keys():  # Using loop to get all of the grouped data and labels of grades
        labels_list.append(grade)
        df_list.append(grouped_df.get_group(grade)["price"])
plt.xlabel("grade")   # Adding label of x variable
plt.ylabel("price") # Adding label of y variable
plt.ticklabel_format(style='plain') # Removing scientific notation (1e6)
plt.boxplot(df_list, labels=labels_list) # Creating boxplot using boxplot function
plt.show() # Showing the plot

#Hint: plt.boxplot
```

The boxplot gives more information about medians etc. in comparison to the scatterplot. We can still see the rising tendancy of the price according to the grade, but also it shows us the amount of outliers.

```python
#b)
from statsmodels.graphics.mosaicplot import mosaic

#Your code:
mosaic(df, ['grade', 'condition'], title='grade to condition')
plt.show()
```

Mosaic plot visualise that around half of the houses have condition 3 and grades around 6-11.

```python
#b)
#Your code:
plt.scatter(df[['grade']],df[['price']], c='black', marker=".", label='All')
plt.scatter(df2[['grade']],df2[['price']], c='y', marker="o", s=100, label='Accepted')
plt.legend(loc='upper left')
plt.show()
```
